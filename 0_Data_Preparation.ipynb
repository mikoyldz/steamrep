## Einführung

In diesem Notebook fokussieren wir uns auf die **Datenvorbereitung** für das Steam Sentiment Analysis Projekt. Die Exploratory Data Analysis (EDA) wurde bereits in einem separaten tebook durchgeführt und die wichtigsten Erkenntnisse werden hier direkt angewendet.

### Warum ist die Datenaufbereitung wichtig?

Um Modelle effektiv zu trainieren, ist eine saubere und gut vorbereitete Datengrundlage essenziell. Die Daten müssen so aufbereitet werden, dass sie sowohl die forderungen der Modelle erfüllen als auch die Qualität und Aussagekraft der Vorhersagen maximiert wird.

---

### Zusammenfassung der EDA-Ergebnisse

Basierend auf der EDA haben wir die folgenden Kriterien für die Datenfilterung und Bereinigung festgelegt:

1. **Zeitlicher Rahmen:** Nur Daten aus den Jahren 2020 und 2021 werden verwendet.
2. **Sprache:** Es werden ausschließlich englischsprachige Reviews berücksichtigt.
3. **Hilfreiche Bewertungen:** Maximale Anzahl an 'helpful votes' beträgt 1500.* 
4. **Humor:** Maximale Anzahl an 'funny votes' beträgt ebenfalls 1000.*
5. **Votes:** Mindestens ein Vote für 'helpful' oder 'funny' ist erforderlich.
6. **Spielzeit:** Die Gesamtspielzeit pro Nutzer darf maximal 15.000 Stunden betragen.

**Sehr hohe Votes sind extrem selten und könnten auf Spam/Manipulation hinweisen.*

---

### Zielorientierte Datenbereinigung

Da das Ziel dieses Projekts darin besteht, ein Transformer-Modell zu entwickeln, das automatisch vorhersagen kann, ob ein Review positiv oder negativ ist, werden die folgenden hritte zur Datenbereinigung unternommen:

1. **Textlänge:** Reviews werden auf 512 Tokens begrenzt.
2. **Zu kurze Reviews:** Reviews mit weniger als 10 Tokens werden entfernt.
3. **Zu lange Reviews:** Reviews, die mehr als 512 Tokens haben, werden gekürzt.

---

### Daten-/Modellorientierte Vorbereitungen

Für die Vorbereitung der Daten für die Modelle werden folgende Schritte durchgeführt:

1. **Duplikate entfernen:** Sicherstellen, dass keine doppelten Reviews im Datensatz vorhanden sind.
2. **Sonderzeichen und Emojis:** Transformer-Modelle können mit Sonderzeichen umgehen, daher sind keine speziellen Bereinigungen nötig.
3. **HTML-Tags:** HTML-Tags könnten unnötige Tokens verbrauchen und werden entfernt.
4. **Lowercasing:** Da einige Transformer case-sensitive ist, werden alle Texte in Kleinbuchstaben umgewandelt, um die Konsistenz zu erhöhen.
6. **Optionale Balance:** Gruppierung nach Spiel-ID und entfernen des unteren Quantils, um die Klassenbalance zu verbessern.

## Libraries

```python
import pandas as pd #generische Datenmanipulation
import re # string manipulation

from transformers import DistilBertTokenizer #tokenizers des modells, welches trainiert wird
import matplotlib.pyplot as plt #generische Bibliothek für Visualisierungen
```

## Datenbereinigung

```python
#daten einlesen
df = pd.read_parquet("data/source_data/steam_reviews.parquet")
```

```python
df.info()
```

### EDA Orientierte Bereinigung

Die Daten müssen von neu aufbereitet werden, um die oben genannten Kriterien zu erfüllen. 

Dafür werden die Code-Zellen aus der EDA wiederverwendet und um die notwendigen Schritte erweitert.

```python
# Optimierung Datenformat
df = df.astype({
    'app_name':'string',
    'language':'string',
    'review':'string',
    'app_id': 'int32',
    'review_id': 'int32',
    'votes_helpful': 'int32',
    'votes_funny': 'int32',
    'weighted_vote_score': 'float32',            
    'comment_count': 'int32',
    'author.num_games_owned': 'int32',
    'author.num_reviews': 'int32',
    'author.playtime_forever': 'float32',
    'author.playtime_last_two_weeks': 'float32',
    'author.playtime_at_review': 'float32'
})

#Zeitstempel in Datetime umwandeln
df['timestamp_created'] = pd.to_datetime(df['timestamp_created'], unit='s')
df['timestamp_updated'] = pd.to_datetime(df['timestamp_updated'], unit='s')
df['author.last_played'] = pd.to_datetime(df['author.last_played'], unit='s')
```

```python
# Definiere manuelle Grenzwerte für die Bereinigung der Daten
realistic_limits = {
    'votes_helpful': 1500, # mmaximal 1500 helpful votes
    'votes_funny': 1000, # maximal 1000 funny votes
    'author.playtime_forever': 15000, # Maximale Spielzeit von 15000 Minuten haben
    'timestamp_created': pd.Timestamp(2020, 1, 1), # auf das Jahr 2020 beschränken
}

# Filtere die Daten basierend auf den festgelegten Grenzwerten
df = df[(df['votes_helpful'] <= realistic_limits['votes_helpful']) &
                (df['votes_funny'] <= realistic_limits['votes_funny']) &
                (df['author.playtime_forever'] <= realistic_limits['author.playtime_forever']) &
                (df['timestamp_created'] >= realistic_limits['timestamp_created'])]
```

```python
df.head(5)
```

```python
# duplikate entfernen, auf basis aller daten
df = df.drop_duplicates()
# entferne alle Zeilen, die keine Bewertung enthalten
df = df.dropna(subset=["review"])
```

```python
# API Daten Zusammenführen
df_game_info = pd.read_parquet('data/source_data/game_info.parquet')
df = df.merge(df_game_info, on='app_id', how='left')
```

```python
df.head(5)
```

### Daten-/Modellorientierte Bereinigung

```python
# Imbalance der Recommendations untersuchen
label_distribution = df['recommended'].value_counts() / len(df)
print(label_distribution)
```

Zuerst werden alle Reviews, wo kein Vote abgegeben wurde, entfernt.

```python
# sotieren nach votes_helpful und votes_funny und entfernen von Zeilen, die keine votes haben
df = df.sort_values(by=['votes_helpful', 'votes_funny'], ascending=False)
df = df[(df['votes_helpful'] > 0) | (df['votes_funny'] > 0)]
```

```python
df.head(5)
```

Nur 6% der Reviews sind derzeit als not recommended markiert. Wie in der Einleitung schon erwähnt, werden die Daten nach Spiel-ID gruppiert und das untere Quantil entfernen. Dies sollte die Klassenbalance verbessern und die Modelle effektiver machen, da spiele mit einer höheren Anzahl an Reviews wahrscheinlicher auch eine höhere Anzahl an negative Bewertungen haben.

```python
# Gruppiere nach Spiel-ID und Zählen der Bewertungen per ID
reviews_per_game = df['app_id'].value_counts()

# Berechne des unteren Quantils 
lower_bound = reviews_per_game.quantile(0.25)

# Filters für das untere Quantil 
filtered_games = reviews_per_game[(reviews_per_game >= lower_bound)].index
```

```python
# Entferne unteres Quantil
df_filtered = df[df['app_id'].isin(filtered_games)].copy()
df_filtered.head()
```

Nun werden die HTML-Tags entfernt und der Text in Kleinbuchstaben umgewandelt. HTML- Tags könnten unnötige Tokens verbrauchen und die Konsistenz des Textes beeinträchtigen und lowercase ist eine voraussetzung für die meisten Transformer-Modelle.

```python
# Substitution von generischen HTML-Tags und string lowercasting
df_filtered.loc[:,'cleaned_review'] = df_filtered['review'].apply(lambda x: re.sub(r'<.*?>', '', x))
df_filtered['cleaned_review'] = df_filtered['cleaned_review'].str.lower()
```

```python
df_filtered.head(5)
```

Nun werden die Reviews auf eine maximale Länge von 512 Tokens begrenzt und zu kurze Reviews (<10 Tokens) entfernt. Bevor wir das machen, werden wir uns die Token Range der Reviews anschauen, um zu sehen, wie viele Reviews betroffen sind und wie die grundlegende Verteilung aussieht.

```python
tokenizer = DistilBertTokenizer.from_pretrained('distilbert-base-uncased') #einlesen des Tokenizers
```

```python
# Tokenisierung und Zählen der Token-Länge per Review
df_filtered['token_length'] = df_filtered['cleaned_review'].apply(
    lambda x: len(tokenizer.tokenize(x))
)
```

```python
# Visualisierung der Tokenlängenverteilung setz ein xlim und ylim, um die Verteilung besser zu sehen
plt.figure(figsize=(10, 6))
plt.hist(df_filtered['token_length'], bins=1000, color='skyblue', edgecolor='black')
plt.title('Verteilung der Tokenlängen in den Reviews')
plt.xlabel('Tokenanzahl')
plt.ylabel('Anzahl der Reviews')
plt.xlim(0, 512)
plt.grid(True)
plt.show()
```

```python
df_filtered['token_length'].describe()
```

Was uns direkt auffällt ist die Verteilung der Tokens in den Reviews. Die meisten Reviews haben zwischen 10 und 100 Tokens. Es gibt jedoch auch einige Reviews mit mehr als 1000 Tokens. 

```python
# entferne alle reviews mit mehr als 512 tokens und weniger als 10 tokens
df_filtered = df_filtered[(df_filtered['token_length'] <= 512) & (df_filtered['token_length'] >= 10)]
```

```python
# Visualisierung der Tokenlängenverteilung setz ein xlim und ylim, um die Verteilung besser zu sehen
plt.figure(figsize=(10, 6))
plt.hist(df_filtered['token_length'], bins=100, color='skyblue', edgecolor='black')
plt.title('Verteilung der Tokenlängen in den Reviews')
plt.xlabel('Tokenanzahl')
plt.ylabel('Anzahl der Reviews')
plt.xlim(0, 512)
plt.grid(True)
plt.show()
```

```python
df_filtered['token_length'].describe()
```

Wir sehen, dass die Tokenlängenverteilung nach der Bereinigung der Daten deutlich besser aussieht. 
Die meisten Tokenlängen liegen nun zwischen 10 und 100. 
Dies könnte später hilfreich sein, wenn wir mehr Informationen über die Reviews mitgeben wollen, wie das Spiel, die Genre etc.

```python
label_distribution = df_filtered['recommended'].value_counts() / len(df_filtered)
print(label_distribution)
```
